## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
print(df)
```


# Create basic charts

- Line chart
- Bar chart
- Histogram
- Box plot
- Scatter plot




```python
#Line chart
df['mpg'].plot.line(color='purple')
mpl.pyplot.xticks(rotation = 45)
```

```python
#Bar chart
df['mpg'].plot.barh(color='red', xticks=[0,10,20])
```

```python
#Histogram
df['mpg'].plot.hist(bins=15,title='Miles Per Gallon')
```

```python
#Boxplot
df['mpg'].plot.box(color ='red', ylabel = 'test', title = 'this is my chart')
```

```python
#Scatter plot
df.plot.scatter(x='mpg',y='hp',c='wt')
```

```python
#Your code goes here
def autopct_format(values):
        def my_format(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{:.1f}%\n({v:d})'.format(pct, v=val)
        return my_format
transmission = df['am'].value_counts()
mpl.pyplot.pie(transmission, labels=['Automatic', 'Manual'], autopct=autopct_format(transmission), colors = ['lightblue', 'lightgreen'])
mpl.pyplot.title('Transmission Type Distribution')
mpl.pyplot.show()
```
