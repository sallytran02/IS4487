## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Additional Data Preparation

```python
df_clean = df_clean[df_clean['reported_satisfaction'] != 'low']
df_clean.head(10)

```

```python
df_clean = df_clean[(df_clean['college'] == 'one')]
df_clean.tail(10)
```

## Data Visualization

 Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
import matplotlib.pyplot as plt

leave_counts = df_clean['leave'].value_counts()
leave_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Leave Categories')
plt.xlabel('Leave Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(df_clean['income_k'], df_clean['house'], alpha=0.5)
plt.title('Relationship between Income and House Price')
plt.xlabel('Income (in K)')
plt.ylabel('House Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plot to show the distribution of 'Income' by 'Leave' category
plt.figure(figsize=(10, 6))
sns.boxplot(x='leave', y='income_k', data=df_clean, palette='Set2')
plt.title('Distribution of Income by Leave Category')
plt.xlabel('Leave Category')
plt.ylabel('Income (in K)')
plt.show()
```
