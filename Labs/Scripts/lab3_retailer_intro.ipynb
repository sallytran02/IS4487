## Prepare Data

➡️ Assignment Tasks
- Convert the Total_Amount into a number (not a string)
- Remove rows with empty (NULL) values
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
#turn total_amount into numbers not a string
df['total_amount'] = pd.to_numeric(df['total_amount'].replace('[\$,]', '', regex=True))

#remove rows with null values
df = df.dropna()

df.head()



```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

```python
#average price per month per year

df['product_price'] = pd.to_numeric(df['product_price'].replace('[\$,]', '', regex=True), errors='coerce')

avg_year_df = df[['order_year_month', 'product_price']]

avg_year_df = avg_year_df.groupby('order_year_month')['product_price'].mean()

avg_year_df.plot(kind='bar', figsize=(10, 6), color='skyblue')
mpl.pyplot.title('Average Product Prices by Year and Month')
mpl.pyplot.xlabel('Year and Month')
mpl.pyplot.ylabel('Average Product Price')
mpl.pyplot.show()

#average price per month

avg_month_df = df[['order_month', 'product_price']]

avg_month_df = avg_month_df.groupby('order_month')['product_price'].mean()

avg_month_df.plot(kind='bar', figsize=(10,6), color='lightgreen')
mpl.pyplot.title('Average Product Prices by Month')
mpl.pyplot.xlabel('Month')
mpl.pyplot.ylabel('Average Product Price')
mpl.pyplot.show()
```

```python
#total revenue

total_rev_df = df[['order_month', 'total_amount']]

month_rev_df = total_rev_df.groupby('order_month')['total_amount'].sum()

month_rev_df.plot(kind='bar', figsize=(10,6), color='darkgreen')
mpl.pyplot.title('Total Revenue by Month')
mpl.pyplot.xlabel('Month')
mpl.pyplot.ylabel('Total Revenue')
mpl.pyplot.show()
```
